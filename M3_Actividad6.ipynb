```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelBinarizer

df = pd.read_csv('Titanic.csv')
df.head()
```

```python
df2 = pd.DataFrame(df, columns = ['Sex'])
print(df2.head())
```

```python
df3 = pd.get_dummies(df2.Sex, prefix = 'Sex')
df3.head()
```

```python
y = LabelBinarizer().fit_transform(df3)
y
```
